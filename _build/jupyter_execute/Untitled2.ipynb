```python
data = pd.read_csv('Annual_YVR_Temp.csv', parse_dates=['date']) 
data['year'] = pd.DatetimeIndex(data['date']).year
data['month'] = pd.DatetimeIndex(data['date']).month
data.head()
```

```python
data.info()
```

```python
precip_sort = data.sort_values("precip")
precip_sort.tail(11)
```

```python
month_min_precip = data.groupby('month')['total_precip'].min().reset_index().round(2)
month_min_precip
```

```python
month_max_precip = data.groupby('month')['total_precip'].max().reset_index().round(2)
month_max_precip
```

```python
jan_precip = data.loc[data['month'] == 1]
jan_precip.tail(35)
```

```python
years = list(data['year'][::5])

pc_bar = alt.Chart(data).mark_bar().encode(
    x=alt.X('year:O', title=None),
    y=alt.Y('precip:Q', title='Precipitation (mm)', axis=alt.Axis(values=years, labelAngle=-90, grid=False)),
    color=alt.Color('precip:Q', scale=alt.Scale(scheme='viridis', reverse=True), legend=None),
    tooltip=[alt.Tooltip('year:O', title="Year"), alt.Tooltip('precip:Q', title="Precip")]
).properties(width=800, title='Total Annual Precipitation - Vancouver International Airport')
pc_bar
```
